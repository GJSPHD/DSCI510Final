```python
import pandas as pd

# List of dataset variables
datasets = {
    "gps": df_gps,
    "ptsd": df_ptsd,
    "LIWC": df_LIWC,
    "communication": df_communication,
    "trauma": df_trauma,
    "tobacco": df_tobacco,
    "support": df_support,
    "su": df_su,
    "somatic": df_somatic,
    "snsis": df_snsis,
    "sleep": df_sleep,
    "ses": df_ses,
    "risk": df_risk,
    "recovery": df_recovery,
    "pss": df_pss,
    "promis_see": df_promis_see,
    "polytrauma": df_polytrauma,
    "personality": df_personality,
    "pdi_mod": df_pdi_mod,
    "pain_mod": df_pain_mod,
    "medication": df_medication,
    "mceps_mod": df_mceps_mod,
    "lte_mod": df_lte_mod,
    "lhcidi_mod": df_lhcidi_mod,
    "injury_mod": df_injury_mod,
    "impulsivebehavior_mod": df_impulsivebehavior_mod,
    "hsu_mod": df_hsu_mod,
    "general_mod": df_general_mod,
    "gen_health_mod": df_gen_health_mod,
    "finimpact_mod": df_finimpact_mod,
    "ffmq_mod": df_ffmq_mod,
    "dissociation_mod": df_dissociation_mod,
    "disability_mod": df_disability_mod,
    "depression_mod": df_depression_mod,
    "demogr_mod": df_demogr_mod,
    "ctq_mod": df_ctq_mod,
    "cs_mod": df_cs_mod,
    "collision_mod": df_collision_mod,
    "cdrs_mod": df_cdrs_mod,
    "bfi_mod": df_bfi_mod,
    "anxiety_mod": df_anxiety_mod,
    "adult_adhd_mod": df_adult_adhd_mod,
}

# Create an empty DataFrame to store results
pid_summary = pd.DataFrame()

# Iterate over datasets and count PIDs
for name, df in datasets.items():
    if "PID" in df.columns:
        pid_counts = df["PID"].value_counts().reset_index()
        pid_counts.columns = ["PID", name]
        if pid_summary.empty:
            pid_summary = pid_counts
        else:
            pid_summary = pd.merge(pid_summary, pid_counts, on="PID", how="outer")

# Fill NaN values with 0 and convert counts to integers
pid_summary = pid_summary.fillna(0).astype(int)

# Save the results to a CSV file
pid_summary.to_csv("PID_Counts_Per_Dataset.csv", index=False)

# Print a message for the user
print("PID counts per dataset saved as 'PID_Counts_Per_Dataset.csv'.")
```

```python
import pandas as pd
import itertools
import seaborn as sns
import matplotlib.pyplot as plt

# Define the short-format datasets in a dictionary
short_format_dfs = {
    "ptsd": df_ptsd,
    "trauma": df_trauma,
    "tobacco": df_tobacco,
    "support": df_support,
    "su": df_su,
    "somatic": df_somatic,
    "snsis": df_snsis,
    "sleep": df_sleep,
    "ses": df_ses,
    "risk": df_risk,
    "recovery": df_recovery,
    "pss": df_pss,
    "promis_see": df_promis_see,
    "polytrauma": df_polytrauma,
    "personality": df_personality,
    "pdi_mod": df_pdi_mod,
    "pain_mod": df_pain_mod,
    "medication": df_medication,
    "mceps_mod": df_mceps_mod,
    "lte_mod": df_lte_mod,
    "lhcidi_mod": df_lhcidi_mod,
    "injury_mod": df_injury_mod,
    "impulsivebehavior_mod": df_impulsivebehavior_mod,
    "hsu_mod": df_hsu_mod,
    "general_mod": df_general_mod,
    "gen_health_mod": df_gen_health_mod,
    "finimpact_mod": df_finimpact_mod,
    "ffmq_mod": df_ffmq_mod,
    "dissociation_mod": df_dissociation_mod,
    "disability_mod": df_disability_mod,
    "depression_mod": df_depression_mod,
    "demogr_mod": df_demogr_mod,
    "ctq_mod": df_ctq_mod,
    "cs_mod": df_cs_mod,
    "collision_mod": df_collision_mod,
    "cdrs_mod": df_cdrs_mod,
    "bfi_mod": df_bfi_mod,
    "anxiety_mod": df_anxiety_mod,
    "adult_adhd_mod": df_adult_adhd_mod
}

# Extract unique PIDs from each dataset
pid_dict = {name: set(df["PID"].dropna().astype(str).unique()) for name, df in short_format_dfs.items()}

# Create an empty DataFrame to store PID overlaps
pid_overlap_matrix = pd.DataFrame(index=pid_dict.keys(), columns=pid_dict.keys(), dtype=float)

# Compute pairwise intersections
for (df1_name, df2_name) in itertools.combinations(pid_dict.keys(), 2):
    intersection_size = len(pid_dict[df1_name].intersection(pid_dict[df2_name]))
    pid_overlap_matrix.loc[df1_name, df2_name] = intersection_size
    pid_overlap_matrix.loc[df2_name, df1_name] = intersection_size  # Symmetric matrix

# Fill diagonal with total unique PIDs per dataset
for dataset in pid_dict.keys():
    pid_overlap_matrix.loc[dataset, dataset] = len(pid_dict[dataset])

# Convert to numeric
pid_overlap_matrix = pid_overlap_matrix.astype(float)

# Plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(pid_overlap_matrix, annot=True, fmt=".0f", cmap="Blues", linewidths=0.5)
plt.title("Pairwise PID Overlaps Between Short-Format Datasets")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
